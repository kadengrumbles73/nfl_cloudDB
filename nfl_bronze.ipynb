```python
%sql
-- Database for the raw data
CREATE SCHEMA IF NOT EXISTS workspace.nfl;
-- raw files volume
CREATE VOLUME IF NOT EXISTS workspace.nfl.bronze_landing;
```

```python
import requests
import json
from datetime import datetime
import time
  
# Gets all teams
url = "https://site.api.espn.com/apis/site/v2/sports/football/nfl/teams"

# destination path
output_path = "/Volumes/workspace/nfl/bronze_landing/"
file_name = f"nfl_teams_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
# doesent overwite the files already in the volume we want each days data cus it can change

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    # Save to the Volume
    with open(f"{output_path}{file_name}", "w") as f:
        json.dump(data, f)
```

```python
# get last 3 seasons
seasons = [2023, 2024, 2025]
output_path = "/Volumes/workspace/nfl/bronze_landing/"


for season in seasons:
    # get all  18 weeks of the NFL regular season
    for week in range(1, 19):
        
        # write specific link for each week
        url = f"https://site.api.espn.com/apis/site/v2/sports/football/nfl/scoreboard?dates={season}&week={week}"
        
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            
            # Label the file with the season and week
            file_name = f"nfl_schedule_{season}_week{week}.json"
            
            with open(f"{output_path}{file_name}", "w") as f:
                json.dump(data,  f)
            
            # sleep for a lil bit
            time.sleep(0.5)
```
